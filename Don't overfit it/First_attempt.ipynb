```python
import pandas as pd
import numpy as np

np.random.seed(seed = 12)
```

```python
test_set = pd.read_csv("./test.csv")
train_set = pd.read_csv("./train.csv")
```

```python
test_set.shape, train_set.shape
```

```python
train_set.head(10)
```

```python
train_features, labels = train_set.drop(["target", "id"], axis=1), train_set.target.values
```

```python
train_features.shape, labels.shape
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression(C=1, solver='liblinear')
```

```python
logit.fit(train_features, labels)
```

```python
def write_to_submission_file(predicted_labels, out_file, test_set,
                             target='target', index_label="id"):
    predicted_df = pd.DataFrame(predicted_labels,
                                index = test_set["id"],
                                columns=[target])
    predicted_df.to_csv(out_file, index_label=index_label)
```

```python
logit_test_pred = logit.predict(test_set.drop(["id"], axis=1))
write_to_submission_file(logit_test_pred, 'subm1.csv', test_set=test_set)
```
